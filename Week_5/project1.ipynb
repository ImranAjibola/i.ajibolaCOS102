```python
import pandas as pd
df = pd.read_csv(r"C:\Users\Imraan\OneDrive\Documents\i.ajibolaCOS102\Week_5\GIG-logistics.csv")
```

```python
df["Username"] = df["SURNAME"] + " " + df["FIRST NAME"]
```

```python
import tkinter as tk
from tkinter import messagebox

def welcome_employee(username, department):
    department_employees = df[df["DEPARTMENT"] == department]["Username"].tolist()

    if username in department_employees:
        messagebox.showinfo("Welcome", f"Welcome {username}!\nOther members of {department} department: {', '.join(department_employees)}")
    else:
        messagebox.showerror("Error", "Employee not found.")

def submit():
    username = username_entry.get()
    department = department_entry.get()
    welcome_employee(username, department)

root = tk.Tk()
root.title("Employee Verification")
root.geometry("400x200")

username_label = tk.Label(root, text="Username:")
username_label.pack()
username_entry = tk.Entry(root)
username_entry.pack()

department_label = tk.Label(root, text="Department:")
department_label.pack()
department_entry = tk.Entry(root)
department_entry.pack()

submit_button = tk.Button(root, text="Submit", command=submit)
submit_button.pack()

root.mainloop()
```
